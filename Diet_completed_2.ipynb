```python
def error(list_food, train_percent, k):
    total_error_list = []
    x = 0
    df = pd.concat(list_food)
    print(df.shape)
    
    for s in train_percent:
        for k_val in k:
            train = int(( train_percent[x] *24 )/100)
            test = 24 - train
            col1_train = 3
            col2_train = train + 3
            col1_test = train + 3
            col2_test = 27

            # finds euclidean distance / similarity
            def euclidean(x, y):
                distances = np.sqrt(np.sum(np.square(y - x), axis=1))
                return distances
            calculations = [] 
            indexes = list(df.index)
            # finding similarity / distance from 1 element to all the other elements
            for i in indexes:
                selected1 = df.loc[df["description"]==df["description"][i]]
                selected = selected1[selected1.columns[col1_train:col2_train]]
                remaining = df[df.columns[col1_train:col2_train]]
                selected_test = selected1[selected1.columns[col1_test:col2_test]]
                remaining_test = df[df.columns[col1_test:col2_test]]
                selected = np.array(selected)
                remaining = np.array(remaining)
                z = euclidean(selected, remaining)
                calculations.append(z)

            # list of k nearest neighbors
            list3 = []
            for i in range(len(df)):
                list3.append(sorted(list(enumerate(calculations[i])), key = lambda x:x[1])[0:k_val+1]) 
            
            # Finding percentage error
            error_list = []
            error = 0
            rms = 0
            mean_absolute = 0
            mean_abs2 = 0
            MS = 0
            for i in range(len(df)):
                list2 = pd.DataFrame(list3[i])
                index_of_elem = list2[0].tolist()
                data_test = data_norm[data_norm.columns[col1_test:col2_test]]
                
                # forming a dataset of K nearest elements:
                new2 = []
                for z in index_of_elem:
                    new2.append(data_test.iloc[z])
                selected_D_test = pd.DataFrame(new2)
                avg = selected_D_test.mean()
                actual = data_norm[data_norm.columns[col1_test:col2_test]]
                act = actual.iloc[i]
                act_mean = act.mean()
                avg_mean = avg.mean()
                first = abs(act - avg)
                second = abs(act_mean - avg_mean)
                mean_abs2 += first
                mean_absolute += second
                square = first**2
                rms += square
                sqaured_mean = square.mean()
                MS += sqaured_mean
                err = first.mean()
                error += err
                error_list.append(err)
            rms_mean = rms.mean()
            RMS = math.sqrt(rms_mean)
            error_mean = (error/len(df))*100
            MAE_1 = mean_absolute/len(df)
            MAE_2 = mean_abs2.mean()
            Mean_sq_err = (MS/len(df))*100

            print("The total error for k = {} with training split of {}% is : {:.2f}%".format(k_val,s,error_mean))
#             print("Mean Square Error 1: {:.2f}".format(rms_mean))
#             print("Mean Square Error 2: {:.2f}%".format(Mean_sq_err))
#             print("Root Mean squared error = {:.2f}".format(RMS))
#             print("Mean absolute error_1 = {:.2f}".format(MAE_1))
#             print("Mean absolute error_2 = {:.2f}".format(MAE_2))
            print("------------------------")
        x+=1
    return list3
```

```python
def recommend_1(k):
    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.neighbors import NearestNeighbors
    
    k_value = k
    
    tfidf = TfidfVectorizer(stop_words="english")
    vectors = tfidf.fit_transform(data["description"]).toarray()
    knn = NearestNeighbors(n_neighbors=k_value, algorithm='brute', metric='cosine')
    knn.fit(vectors)


    def recommend(index):
        calories = []
        energy = 0
        remaining = cal_per_cat
        # Finding the index of the input recipe
        recipe_index = index
        _, indices = knn.kneighbors(vectors[recipe_index].reshape(1,-1))
        # Printing only the top k most similar recipes
        for i in indices[0][1:15]:
            if energy < cal_per_cat:
                if data.iloc[i].energy_kcal < remaining: 
                    print("{}, -- calories = {} Kcal".format(data.iloc[i].description, data.iloc[i].energy_kcal))
                    calories.append(data.iloc[i].energy_kcal)
                    energy += data.iloc[i].energy_kcal
                    remaining = cal_per_cat - energy
        print("\nTotal calorie: {} Kcal".format(energy))
        print("--------------------")
        return np.sum(calories)


    # Testing the recommendation function
    total_cal = []
    print("THE RECOMMENDED FOODS ARE\n")
    for item in final:
        energy = recommend(item)
        total_cal.append(energy)

    total_energy = np.sum(total_cal)
    error = abs(total_energy - cal)/cal
#     print("error",error)
    b = i
    while error > .15:
        b+=1
        print("{}, -- calories = {} Kcal".format(data.iloc[b].description, data.iloc[b].energy_kcal))
        total_energy += data.iloc[b].energy_kcal
        error = abs(total_energy - cal)/cal
    print("-------------")
    print("The total calories of the recommended food ",total_energy,"Kcal")
    print("Error percentage between calories actually needed and recommended calories = {:.2f}%".format(error*100))
    
     
```

```python

def recommend_2(cal_per_categ,split,k):
    
    def euclidean(x, y):
        distances = np.sqrt(np.sum(np.square(y - x), axis=1))
        return distances
    # 
    list5 = []
    for x in range(len(final)):
        for y in final[x]:
            list5.append(y)
    #
    total_energy = []
    for item_index in list5:
        remain = cal_per_categ
        total_cal = 0
        energy = 0
        x = item_index
        select = data.loc[data["description"]==data["description"][x]]
        selected = select[select.columns[3:split+3]]
        sel = np.array(selected)
        remaining = data[data.columns[3:split+3]]
        re = np.array(remaining)
        distance = euclidean(sel, re)
        # recommending top K 
        recommend = sorted(list(enumerate(distance)), key = lambda x:x[1])[0:k]
        for i in range(len(recommend)):
            x = int(recommend[i][0])
            food = data.iloc[x]
            if energy < cal_per_categ:
                if int(food.energy_kcal) < remain:
                    print("{}, -- calories = {} Kcal".format(food.description, food.energy_kcal))
                    energy += int(food.energy_kcal)
                    total_cal = energy
                    remain = cal_per_categ - energy

        total_energy.append(total_cal)
        print("\nTotal calorie: {} Kcal".format(total_cal))
        print("------------------")

    
    total_ = np.sum(total_energy)
    error = abs(total_ - cal)/cal
#     print("error",error)
    b = x
    while error > .15:
        b+=1
        food = data.iloc[b]
        print("{}, -- calories = {} Kcal".format(food.description, food.energy_kcal))
        total_+= food.energy_kcal
        error = abs(total_ - cal)/cal
    print("----------")
    print("The total calories of the recommended food ",total_,"Kcal")
    print("Error percentage between calories actually needed and recommended calories = {:.2f}%".format(error*100))
```
